```python
## Spacy and NLTK FOR WORD TOKERNIZATION
```

```python
from nltk.tokenize import sent_tokenize
```

```python
corpus= """Hello I'm giriselvan. I love prranjali. But she is comiited to another boy.I dont know why. She look like shit.But only she left in my thoughts good things are happening in my life.
I am happy!!!!! now really. My life is very great now this point."""
```

```python
print(corpus)
```

```python
documnets= sent_tokenize(corpus)
```

```python
type(documnets)
```

```python
for sentences in documnets:
    print(sentences)
```

```python
## Paragarphs -> Words
## Sentence -> words


from nltk.tokenize import word_tokenize

```

```python
word_tokenize(corpus)
```

```python
for i in documnets:
    print(word_tokenize(i))
```

```python
from nltk.tokenize import wordpunct_tokenize

wordpunct_tokenize(corpus)
```

```python
from nltk.tokenize import TreebankWordTokenizer
tokeneizer = TreebankWordTokenizer()
tokeneizer.tokenize(corpus)

```
