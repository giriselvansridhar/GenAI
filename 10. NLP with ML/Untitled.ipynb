```python
from nltk.stem import PorterStemmer

# Initialize the stemmer
stemming = PorterStemmer()

# List of words to stem
WORDS = ["eating", "eat", "eaten", "writes", "writing"]

# Iterate and print the stemmed words
for word in WORDS:
    print(word + " ---> " + stemming.stem(word))
```

```python
from nltk.stem import RegexpStemmer
```

```python
reg_stemmer= RegexpStemmer('ing$|s$|e$|able$', min = 4)
```

```python
from nltk.stem import SnowballStemmer

# Create an instance of the SnowballStemmer for English
stemmer = SnowballStemmer("english")

WORDS = ["running", "flies", "better", "happiness"]

# Iterate through the words and stem them
for word in WORDS:
    print(word + " --------> " + stemmer.stem(word))
```
