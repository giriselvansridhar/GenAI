```python
def count_substring(string, sub_string):
    Sub_String_First_Letter= sub_string[0]
    Length_of_string=len(string)
    Length_of_sub_string=len(sub_string)
    count=0
    
    for i in range(0, Length_of_string):
        Conition=False
        if Sub_String_First_Letter == string[i]:
            
            for j in range(1, Length_of_sub_string):
                    if string[i+j] == sub_string[j]:
                        Conition = True
                    else: 
                        Conition = False  
        if Conition == True:
                       count+=1
    return count                      

if __name__ == '__main__':
    string = input().strip()
    sub_string = input().strip()
    
    count = count_substring(string, sub_string)
    print(count)
```
