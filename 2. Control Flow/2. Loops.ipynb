```python
n=10
sum=0
for i in range(n):
    sum=sum+i

print(sum)    
```

```python
## Calculate the sum of first n natural number using While and FOR loop

n= 10
sum = 0
count=1
while count<=n:
    sum=sum+count
    count=count+1
print("Sum of the first 10 natural number", sum)
```
