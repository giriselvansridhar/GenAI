```python
## Variable Length Argumnets
## Positional and Keyword arguments


def print_numbers(*krish):
    for numbers in krish:
        print(numbers)

print_numbers(1,2,2,3,4,5,5,6,6,7,"Giri")
```

```python
### Keyword Argumnets

def print_details(*args,**kwargs):
    for val in args:
        print("Positonal ARGUMNENT: "+ str(val))
    for key, value in kwargs.items():
        print(f"{key}: {value}")



```

```python
print_details(1,2,3,4,"Krish", name="Krish", age="32", country="Indis")
```
